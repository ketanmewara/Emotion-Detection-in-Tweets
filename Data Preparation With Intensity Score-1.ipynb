# Data Cleaning

```python
# Remove Unwanted text pattern and word from the tweets
def remove_text(input_txt, pattern):
    r = re.findall(pattern,input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt
```

```python
import numpy as np
import re
```

```python
# 1. Remove @user
dataset['clean_tweet'] = np.vectorize(remove_text)(dataset['Tweet'],"@[\w]*")
```

```python
dataset.head()
```

```python
dataset['Tweet'][2]
```

```python
dataset['clean_tweet'][2]
```

```python
# 2. Remove Punctuation, Numbers, Special Charaters
dataset['clean_tweet'] = dataset['clean_tweet'].str.replace("[^a-zA-Z#]"," ")
```

```python
# 3. Remove Short Words like: hmm, oh etc
dataset['clean_tweet'] = dataset['clean_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
dataset.drop(['Tweet'],axis=1, inplace=True)
```

# Data Preprocessing

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
wordnet = WordNetLemmatizer()
ps = PorterStemmer()
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing

corpus = []
for i in range(0, len(dataset)):
# #     print(i)
    tweets = dataset['clean_tweet'][i]  
    tweets = tweets.lower()
    tweets = tweets.split()
    
    tweets = [wordnet.lemmatize(word) for word in tweets if not word in stopwords.words('english')]
    tweets = ' '.join(tweets)
    corpus.append(tweets)
```

```python
corpus
```

# Model Building

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM,Conv1D,GlobalMaxPool1D,TimeDistributed
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Bidirectional
```

```python
tf.__version__
```

```python
### Vocabulary size
vocab_size=5000

### onehot representation
onehot_repr=[one_hot(words,vocab_size)for words in corpus] 
onehot_repr[:10]
```

```python
sent_length=280
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
len(embedded_docs),dataset['Classes'].shape
```

```python
train = embedded_docs
target = dataset['Classes']
train.shape, target.shape
```

```python
target
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
labels = dataset['Classes'].replace({'anger':0,'fear':1,'joy':2,'sadness':3})
```

```python
labels
```

```python
labels = to_categorical(labels,num_classes=4)
labels
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
emb_dim = 100
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, labels, test_size=0.33, random_state=94)
```

```python
def model():
    # Creating model
    modelA=Sequential()
    modelA.add(Embedding(vocab_size, emb_dim, input_length=embedded_docs.shape[1]))
    modelA.add(Conv1D(128, kernel_size=3, activation = "relu"))
    modelA.add(GlobalMaxPool1D())
    modelA.add(Dense(4,activation='softmax'))
    modelA.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    return modelA
```

```python
model = model()
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=32,verbose=1)
```

```python
model.evaluate(X_test,y_test)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Train vs Validation Accuracy')
plt.ylabel('Acc')
plt.xlabel('epoch')
plt.legend(['train','validation'])
plt.show()
```
